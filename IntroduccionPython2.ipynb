# Usando los datos

## Dividir un dataset (Entrenamiento y Test)

```python
iris_dataset["variety"].unique()
```

```python
len(iris_dataset[iris_dataset["variety"] == "Setosa"].sample(frac=0.20))
```

```python
test_iris = pd.DataFrame()

for species in iris_dataset["variety"].unique():
    test_iris = pd.concat([test_iris, iris_dataset[iris_dataset["variety"] == species].sample(frac=0.20)])

test_iris
```

```python
len(test_iris)
```

```python
iris_dataset.index.isin(test_iris.index)
```

```python
train_iris = iris_dataset[~iris_dataset.index.isin(test_iris.index)]
train_iris
```

```python
test_iris
```

```python
train_iris.reset_index(inplace=True, drop=True)
test_iris.reset_index(inplace=True, drop=True)
```
